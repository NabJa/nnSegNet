# Scrape Result table from Synapse

Result table at https://www.synapse.org/#!Synapse:syn25829067/wiki/610869

```python
from bs4 import BeautifulSoup
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd
import re
```

## Use selenium chrome browser to webscrape Synapse result table
Simple requests are unnecessary complicated due to two reasons:
1. Javascript execution is needed
2. Default *strict-origin-when-cross-origin* policy

```python
# Must be installed from https://chromedriver.chromium.org/
DRIVER_PATH = r"C:\Users\Jabareen\Downloads\chromedriver"
```

```python
# Set browser to headless mode
options = Options()
options.headless = False # Does not work in headless mode for some reason...
```

```python
# Run a browser
driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
```

```python
# Request url containing result table
driver.get("https://www.synapse.org/#!Synapse:syn25829067/wiki/610869")
```

```python
# Wait until table is loaded!!
# _ = driver.page_source
# sleep(8)
```

```python
# Read tables
dfs = pd.read_html(driver.page_source)
```

```python
# Close chrome driver
driver.quit()
```

## Filter table

```python
# Last dataframe contains relevant informations
df = dfs[-1]
```

```python
filtered = df.groupby("Participant/Team").max()
```
